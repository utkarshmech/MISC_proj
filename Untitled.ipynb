```python
# import required libraries
import numpy as np
import matplotlib.pyplot as pl
```

```python
# definte the problem parameters
L=1
nx=20
nt=4000
```

**dif_func** is the function that will actually solve the Forward Finite Difference method. 
- Put bc =0 for question 1 
- and bc= 1 for question 2. (By default bc is taken 0.)

```python
def dif_func(D, L,nx,nt,T, bc=0):
    x = np.linspace(0, L, nx)    # mesh points in space
    dx = x[1] - x[0]
    t = np.linspace(0, T, nt)    # mesh points in time
    dt = t[1] - t[0]
    F = D*dt/dx**2
    # print(F,nt, dt ,dx**2)
    # Assume initially C at all internal points(x) is zero 
    C   = np.zeros(nx)          # unknown C at new time level
    C_1 = np.zeros(nx)          # C at the previous time level
   # Initial condition assuming 0 C inside
    for i in range(nx): 
        C_1[i]=np.exp(-(x[i]**2/.01))

    for n in range(0, nt):
        # Compute C at inner mesh points
        for i in range(1, nx-1):
            C[i] = C_1[i] + F*(C_1[i-1] - 2*C_1[i] + C_1[i+1])
        C[0] = 1
        if(bc): C[nx-1] = C[nx-2]
        # Update C_1 before next step
        C_1[:]= C
    
    return(C)
```

Plot for different D<1, D=1, D>1 as well as bc =1 , bc=0

```python
def splot(bc=0, D=0.5):
    T=(0.01,.1, 1, 10)
    pl.title('BCs from question %i , T in seconds' % (bc+1) ) 
    pl.xlabel("x (distance m)")
    pl.ylabel("C (concentration)")
    for tp in T: pl.plot(np.linspace(0, L, nx), dif_func(D, L ,nx,nt ,tp, bc),label="T=%f" % tp)
    pl.legend()
    return None
```
